```r
CtoR = function(clusts, only_bornes = TRUE, n_sample = 500, return_error = FALSE){
  rep = list()
  law = list()
  local_errors = c()
  for(j in 1:length(clusts)){
    mat = NULL
    law_j = c()
    diss_tot = 0
    for(k in 1:ncol(clusts[[j]])){
      diss_unif = 10^5
      for(delta in c(1/4,1/2,3/4,1)){
       unif_prov = try_unif_1d(clusts[[j]][,k], delta = delta)
       unif_tilde_prov = from_law_to_sample(list(unif_prov), law = list("unif"), n = n_sample)[[1]]
       diss_unif_prov = wasserstein1d(unif_tilde_prov, clusts[[j]][,k],p=2)
       if(diss_unif_prov < diss_unif){
         unif = unif_prov
         unif_tilde = unif_tilde_prov
         diss_unif = diss_unif_prov
       }
      }
     dirac = mean(clusts[[j]][,k])
     dirac = matrix(c(dirac, dirac))
     dirac_tilde = from_law_to_sample(list(dirac), law = list("dirac"), n = n_sample)[[1]]
     diss_dirac = wasserstein1d(dirac_tilde, clusts[[j]][,k],p=2)
     dd <<- unif_tilde
     if(diss_dirac > diss_unif){
      if(!only_bornes){mat = cbind(mat, unif_tilde)}
       else{mat = cbind(mat, unif)}
       law_j = c(law_j, "unif")
       diss_tot = diss_tot + diss_unif^2
     }
     else{
       if(!only_bornes){mat = cbind(mat, dirac_tilde)}
       else{mat = cbind(mat, dirac)}
       law_j = c(law_j, "dirac")
       diss_tot = diss_tot + diss_dirac^2
     }
    }
    rep[[j]] = mat
    law[[j]] = law_j
    local_errors = c(local_errors, sqrt(diss_tot))
  }
  res = list(rep, law)
  if(return_error){res = c(res,  list(local_errors, sapply(clusts, nrow)))}
  return(res)
}
```

```r
create_hybrid_sample = function(n, d){
    weights_1 = round(runif(1, min = 0.15, max = 0.7),2)
    weights_2 = round(runif(1, min = 0.15, max = 1-weights_1-0.15),2)
    weights_3 = (1-weights_1-weights_2)
    weights = c(weights_1,weights_2, weights_3)
    sizes = weights*n
    samp_hybrid = NULL
    all_law = list()
    all_rep = list()
    all_clust = list()
    for(xx in 1:3){
        samp = sobol((10^4*xx+1):(10^4*xx+sizes[xx]), dim = d)
        is_unif = which(sample(c(0,1), size = d, replace = TRUE) == 1)
        size_unif = sample(x= c(0.25,0.5, 0.75,1), size = length(is_unif))
        size_marginal = rep(0, length = d)
        size_marginal[is_unif] = size_unif
        repres = runif(n = d, min = 0, max = 1-size_marginal)
        repres = rbind(repres, repres + size_marginal)
        law = rep("dirac", d)
        law[is_unif] = "unif"
        all_rep[[xx]] = repres
        all_clust[[xx]] =  from_unif_to_sample(samp, repres, law)
        all_law[[xx]] = law
    }
    return(list(clusts = all_clust, rep = all_rep, law = all_law))
}

initialize_algo = function(l,d, multistart = 1){
    starts = list()
    for(start in 1:multistart){
        weights = rep(1/l, l = l)
        samp_hybrid = NULL
        all_law = list()
        all_rep = list()
        for(xx in 1:l){
            size_unif = sample(x= c(0,0.25,0.5, 0.75,1), size = d)
            repres = runif(n = d, min = 0, max = 1-size_unif)
            repres = rbind(repres, repres + size_unif)
            law = rep("dirac", d)
            law[which(size_unif >0.001)] = "unif"
            all_rep[[xx]] = repres
            all_law[[xx]] = law
        }
    starts[[start]] = list(rep = all_rep, law = all_law)
    }
    return(starts)
}



```

```r
path = "df_follow_bis_1.csv"
set.seed(1234)

#set.seed(1)

list_samples = lapply(1:15, function(ii){create_hybrid_sample(n = 200, d = 3)})

true_mixt = list()
aq_mixt = list()
errors_true = list()
errors_aq = list()
evol_errors = list()

write.csv(rbind(c(0,0,0,0)), path, row.names = FALSE)
for(ii in 1:15){
    errors_aq[[ii]] = c(10^5,10^5)
    df_follow = as.matrix(read.csv(path))
    write.csv(rbind(df_follow,c(as.numeric(errors_aq[[ii]]), c(0,0))), path, row.names = FALSE)
    
    samp = list_samples[[ii]]
    init = initialize_algo(l = 3, d = 3, multistart = 2)
    for(start in 1:2){
        res_hybrid = augmented_quanti(samp = do.call(rbind,samp[[1]]), rep = init[[start]]$rep,vec_prop = c(0.4,0.2,0.1), it_lim = 10,n_sample = 1000,threshold = 0.006, law =init[[start]]$law, prop_search = 1)
        best_hybrid = find_best(res_hybrid)
        
        evol_error = get_evol_error(res_hybrid)
        error = compute_errors(best_hybrid[[1]], best_hybrid[[2]], best_hybrid[[3]])
        if(error[[1]] < errors_aq[[ii]][[1]]){
            evol_errors[[ii]] = evol_error
            true_mixt[[ii]] = samp
            aq_mixt[[ii]] = list(best_hybrid[[1]], best_hybrid[[2]], best_hybrid[[3]])
            
            errors_true[[ii]] = compute_errors(samp[[1]],samp[[2]], samp[[3]])
            errors_aq[[ii]] = error
    
            df_follow = as.matrix(read.csv(path))
            df_follow[ii+1,] = c(as.numeric(errors_aq[[ii]]), as.numeric(errors_true[[ii]]))
            write.csv(df_follow, path, row.names = FALSE)
        }

    }
}

list_results = list(true_mixt = true_mixt, aq_mixt = aq_mixt, errors_true = errors_true, errors_aq = errors_aq, evol_errors = evol_errors)
save(list_results,file = "perf_hybrid.RData")
```
